```python
import tensorflow as tf

class AdditiveAttention(tf.keras.layers.Layer):
    def __init__(self, units):
        super(AdditiveAttention, self).__init__()
        self.w1 = tf.keras.layers.Dense(units, activation='linear')  # For keys
        self.w2 = tf.keras.layers.Dense(units, activation='linear')  # For queries
        self.v = tf.keras.layers.Dense(1)  # To calculate alignment scores

    def call(self, query, value):
        transformed_query = self.w2(query)
        transformed_value = self.w1(value)
        alignment_score = self.v(tf.nn.tanh(tf.expand_dims(transformed_query, 1) + transformed_value))
        attention_weights = tf.nn.softmax(alignment_score, axis=1)
        context_vector = tf.reduce_sum(attention_weights * value, axis=1)
        return context_vector, attention_weights
```

```python
batch_size = 2
seq_len = 5 
hidden_dims  = 256
units = 128 

query = tf.random.normal((batch_size , hidden_dims))
value = tf.random.normal((batch_size ,seq_len , hidden_dims))
attention_layer = AdditiveAttention(units)
context_vector , attention_weights = attention_layer(query , value)
```
